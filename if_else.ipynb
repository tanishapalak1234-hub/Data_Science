```python
##if statement-; it is used to check a prticular condition and if the condition satifies then the code inside this block wil work.
##else-: used when the condition inside if fails. 
##elif-: used in the same way as else if. to check multiple conditions.
age=int(input("enter the number"))
if(age>30):
    print("age is greatr than 30")
else:
    print("age is less than 30")
```

```python
##nested loops.
##program to check if a number is positive negative or zero and if positive then is it odd or even.

a=int(input("enter the number"))
if a>0:
    print("the number is positive")
    if a%2==0:
        print("the number is even")
    else:
        print("the number is odd")
elif a<0:
    print("the number is negative")
else:
    print("the number is 0")
```

```python
## simple calculator using nested loops.

a=float(input("enter the number a"))
b=float(input("enter the number b"))
operator=input("+,-,%,/,//,*,**")
if operator=='+':
   add=a+b
   print(add)
elif operator=='-':
   sub=a-b
   print(sub)
elif operator=='*':
   product=a*b
   print(product)
elif operator=='/':
   division=a/b
   print(division)
elif operator=='//':
   if(b!=0):
    floor_division=a//b
    print(floor_division)
   else:
      print("division not possible")
elif operator=='%':
    modulus=a%b
    print(modulus)
elif operator=='**':
   exponent=a**b
   print(exponent)
else:
   print("invalid operator")

```
