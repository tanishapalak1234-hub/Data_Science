```python
##solution 1,2,3 Write a program to check if the provided number is positive or not and if yes then print it out.

n=int(input("enter how many times do you want to check"))

for i in range(0,n):
    n1=int(input("enter the number"))
    if n1==0:
        print("n1 is 0")
    elif n1>0:
        print("n1 is positive")
    else:
        print("n1 is negative")
            
```

```python
## write a program to check if a number is positive and even or positive and odd or negative.

n=int(input("enter how many times you want to check "))

for i in range(0,n):
    n1=int(input("enter the number"))
    if n1>0:
        if n1%2==0:
            print("the given number is positive and even")
        else:
            print("the number is positive and odd")
    elif n1==0:
        print("the number is zero")
    else:
        print('the number is negative')
        
```

```python
## solution 3
n = int(input("Enter a number: "))
sum = 0

while n > 0:
    digit = n % 10      
    sum = sum + digit     
    n = n // 10          

print("Sum of digits =", sum)
```

```python
n = int(input("Enter a number: "))

if n <= 1:
    print("Not a Prime Number")
else:
    for i in range(2, n):
        if n % i == 0:
            print("Not a Prime Number")
            break
    else:
        print("Prime Number")
```

```python
## fibbonacci series.

n = int(input("Enter number of terms: "))

a = 0
b = 1

for i in range(n):
    print(a, end=" ")
    c = a + b
    a = b
    b = c
```
